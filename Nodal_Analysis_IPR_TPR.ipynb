```python
# Data analysis tools

import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Data Source (IPR)

IPR_Q = [0,1999,3094,3902,4512,4963,5275,5458,5519]
IPR_P = [4000,3500,3000,2500,2000,1500,1000,500,14.7]
```

```python
IPR = {"pwf":IPR_P,"flow_rate":IPR_Q}
```

```python
tbl_IPR = pd.DataFrame(IPR)
```

```python
tbl_IPR
```

```python
# Plotting

x_IPR = IPR_Q 
y_IPR = IPR_P

plt.figure(figsize=(8,4))
plt.plot(x_IPR,y_IPR)
plt.xlabel("Flow Rate (bbl/day)")
plt.ylabel("Pressure_Pwf (psi)")
plt.title("IPR Curve")
plt.grid()
```

```python
# Data Source (TPR)

TPR_Q = [1000,1500,2000,2500,3000,3500,4000,4500,5000,5500,6000]

p_190 = [1334,1400,1487,1592,1712,1843,1984,2132,2287,2446,2609]

p_2375 = [1298,1320,1351,1390,1435,1487,1545,1609,1677,1749,1824]

p_2875 = [1286,1294,1305,1319,1336,1356,1378,1403,1431,1461,1493]
```

```python
TPR = {"Flow_Rate": TPR_Q,"Pressure_ts_1.9\"":p_190,"Pressure_ts_2.375\"":p_2375,"Pressure_ts_2.875\"":p_2875}
```

```python
tbl_TPR = pd.DataFrame(TPR)
```

```python
tbl_TPR
```

```python
# Plotting

x_TPR = TPR_Q
y_TPR_190 = p_190
y_TPR_2375 = p_2375
y_TPR_2875 = p_2875

plt.figure(figsize=(8,4))

plt.plot(x_TPR,y_TPR_190, label = "TPR for tubing 1.9\" size" , linewidth= 2)
plt.plot(x_TPR,y_TPR_2375, label = "TPR for tubing 2.375\" size", linewidth= 2)
plt.plot(x_TPR,y_TPR_2875, label = "TPR for tubing 2.875\" size", linewidth= 2)

plt.xlabel("Flow rate (bbl/day)")
plt.ylabel("pressure (Well Head) psi")
plt.grid()
plt.title("TPR Curve")

plt.legend(loc='best')
```

```python
# IPR - TPR Relationship

plt.figure(figsize=(8,4))
plt.style.use("default")

# IPR
plt.plot(x_IPR,y_IPR, label="IPR", linewidth=2)

# TPR
plt.plot(x_TPR,y_TPR_190, label="TPR for tubing 1.9\" size",linewidth=2)
plt.plot(x_TPR,y_TPR_2375, label="TPR for tubing 2.375\" size",linewidth=2)
plt.plot(x_TPR,y_TPR_2875, label="TPR for tubing 2.875\" size",linewidth=2)

# axis label
plt.xlabel("Flow rate (bbl\Day)")
plt.ylabel("FBHP (psi)")
plt.title("Nodal Analysis: IPR-TPR relationship")

plt.legend(loc='best')
plt.grid()
plt.figure(figsize=(8,4))
```
